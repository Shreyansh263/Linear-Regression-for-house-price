```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
n=700
Y=np.zeros(n)
X=np.zeros(n)
```

```python
filename="C:\\Users\\user\\Downloads\\price.csv"       
df=pd.read_csv(filename)
```

```python
j=0
for index,row in df.iterrows():
        Y[j]=row[1]
        X[j]=row[0]
        j+=1
```

```python
plt.scatter(X,Y)
plt.xlabel("Area in square feet")
plt.ylabel("Price in 1000 dollars")
plt.show()
```

Mean square error formula from https://statisticsbyjim.com/regression/mean-squared-error-mse/

```python
w0=0.2
w1= 0.5  #Initial values of weights 
L_rate = 0.0001  #Learning Rate
number_of_iterations = 1000  

#Gradient Descent with number of iterations=1000 
for i in range(number_of_iterations): 
    Y_pred = w1*X + w0  
    sum_1 = (-2/n) * sum(X * (Y - Y_pred))  #Derivative with respect to w0 and w1 respectively
    sum_0 = (-2/n) * sum(Y - Y_pred)  
    w0 = w0-L_rate*sum_0  
    w1 = w1 -L_rate*sum_1  #Updating the values of the weights w0 and w1
    
print (w0, w1)
```

Plotting the regression Line along with the scatter Plot of Y with respect to X.

```python
Y_pred=w1*X+w0

plt.scatter(X,Y) 
plt.plot([min(X),max(X)],[min(Y_pred),max(Y_pred)],color='red')  
plt.xlabel("Area in square feet")
plt.ylabel("Price in 1000 dollars")
plt.show()
```
